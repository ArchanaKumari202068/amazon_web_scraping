```python
#import the necessary libraries:

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# specify the URL of the webpage that we wanted to scrape.
url = "https://www.amazon.in/s?rh=n%3A6612025031&fs=true&ref=lp_6612025031_sar"
```

```python
 # add your user agent to access the html page

HEADERS = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36"}
```

```python
#  Use the requests library to retrieve the HTML content
#Send a HTTP request to the specified URL and save the response from server in a response
response = requests.get(url,headers=HEADERS)
html_content = response.text
```

```python
#  BeautifulSoup object and specify the parser library can be created at the same time. 

soup = BeautifulSoup(html_content, 'html.parser')
```

```python
# show html content
html_content
```

```python
# find_all methodis to find details of all products
products = soup.find_all('div', {'class': "s-card-container"})
```

```python
products
```

```python
# create list to store title,price,rating,availability

title =[]
price =[]
rating=[]
availability=[]

```

```python
# Loop for extracting product details 

for product in products:

    # find() method is used to  finds the first occurrence of the specified value.
    #append() method add a element to the list
    title.append( product.find('span', {'class': "a-text-normal"}).text)
    price.append( product.find('span', {'class': "a-price-whole"}).text)
    rating.append( product.find('span', {'class': "a-size-base"}).text)

    # If the product is unavailable
    try:
        available = product.find('span',{'class':"a-size-small"}).text
    
        
    except:
        available = "Available"
    
    availability.append(available)
    
#     print(title,price,rating)
```

```python
# created a dataframes from the lists
dataframes ={'Product_Name':title,'Price':price,'Rating':rating,'availability_status':availability}
```

```python
dataframes
```

```python
# load data into a DataFrame object:
final_dataframes =pd.DataFrame(dataframes)
print(final_dataframes)
```

```python
#Finally, save all our data in CSV file.
final_dataframes.to_csv("./data.csv")
```
